# Implementing GARCH(1,1)-model to estimate volatilities

- Generalized autoreressive conditional heteroscedasticity (GARCH) is a model to estimate volatities.\
We have chosen to use historical data for NIFTY 50, NIFTY 100, NIFTY 200, NIFTY 500, NIFTY Next 50 and BSE 500 indices.

- In the Indian stock market, these are benchmark stock market indices that represent the weighted average of some of the largest Indian companies listed on the National Stock Exchange (or Bombay stock exchange).\
This is not to be confused with the informal designation "Nifty Fifty" of some of the large-cap stocks on the New York Stock Exchange. 

- There is nothing specific about these stock indices (as opposed to other stock prices) that we use in this project.\
We could do this for other indices or individual stocks.

- We use historical data until the last Friday in August 2024 (which is Aug 30, 2024) to estimate and predict future volatilities.\
Replacing the HistoricalData_*.csv files as more data becomes available will make the predictions more accurate.

```python
import numpy as np
import pandas as pd
import scipy
from matplotlib import pyplot as plt
import seaborn as sns;sns.set()
import warnings
import statsmodels.api as sm

T = 252 #Number of trading days in a year
m = 21 #Number of trading days in a month
```

- We import the historical data (in the form of csv files) and plot the graphs of the values over time.\
The csv files have columns: Index Name, Date, Open, High, Low, Close\
We use data from 1990 to 2024.

- Base period for NIFTY 50 is 3 November, 1995 with base value 1000.\
Base period for NIFTY 100 is January, 2003 with base value 1000.\
Base period for NIFTY 200 is 1 January, 2004 with base value 1000.\
Base period for NIFTY 500 is 3 November, 1995 with base value 1000.\
Base period for NIFTY Next 50 is 1 January, 1995 with base value 1000.\
Base period for BSE 500 is August, 1999 with base value 1000.

- In a year, we have approximately $252$ trading days which is why we have initialized $T$ above to $252$.\
We take a year to have $365$ days and if we subtract $2$ days for the weekend for each of the $52$ weeks, we get $365 - 2\times52 = 365 - 104 = 261$.

- Accounting for stock market being closed on holidays, we have roughly $252$ trading days, each year.\
Consequently, there are roughly $21$ trading days each month.

```python
df50 = pd.read_csv('HistoricalData_Nifty50.csv')
df50.Date = pd.to_datetime(df50.Date)

df100 = pd.read_csv('HistoricalData_Nifty100.csv')
df100.Date = pd.to_datetime(df100.Date)

df200 = pd.read_csv('HistoricalData_Nifty200.csv')
df200.Date = pd.to_datetime(df200.Date)

df500 = pd.read_csv('HistoricalData_Nifty500.csv')
df500.Date = pd.to_datetime(df500.Date)

dfn50 = pd.read_csv('HistoricalData_NiftyNext50.csv')
dfn50.Date = pd.to_datetime(dfn50.Date)

dfb500 = pd.read_csv('HistoricalData_BSE500.csv')
dfb500.Date = pd.to_datetime(dfb500.Date)

plt.figure(figsize=(12, 8))
plt.plot_date(df50.Date, df50.Close, fmt='-', label = 'NIFTY50');
plt.plot_date(df100.Date, df100.Close, fmt='-', label = 'NIFTY100');
plt.plot_date(df200.Date, df200.Close, fmt='-', label = 'NIFTY200');
plt.plot_date(df500.Date, df500.Close, fmt='-', label = 'NIFTY500');
plt.plot_date(dfn50.Date, dfn50.Close, fmt='-', label = 'NIFTYNEXT50');
plt.plot_date(dfb500.Date, dfb500.Close, fmt='-', label = 'BSE500');
plt.legend();
```

- Let $\sigma_n$ denote the volatility of market variable on day $n$ as estimated at the day of $n-1$.\
Then $\sigma_n^2$ is the variance rate.\
We let $S_i$ denote the value of the market variable on day $i$ which in our case is the price of the stock market index.

- The continuously compunded return during day $i$ is what we denote by 'LogReturn' and calculate it as
$$\ln\left(\frac{S_i}{S_{i-1}}\right)$$

- A related quantity that is easier to work with is the 'PercentChange' in the market variable denoted by
$$u_i = \frac{S_i-S_{i-1}}{S_{i-1}}$$

```python
df50['PercentChange'] = df50['Close'].pct_change()
df50['LogReturn'] = np.log(df50['Close'] / df50['Close'].shift(1))

df100['PercentChange'] = df100['Close'].pct_change()
df100['LogReturn'] = np.log(df100['Close'] / df100['Close'].shift(1))

df200['PercentChange'] = df200['Close'].pct_change()
df200['LogReturn'] = np.log(df200['Close'] / df200['Close'].shift(1))

df500['PercentChange'] = df500['Close'].pct_change()
df500['LogReturn'] = np.log(df500['Close'] / df500['Close'].shift(1))

dfn50['PercentChange'] = dfn50['Close'].pct_change()
dfn50['LogReturn'] = np.log(dfn50['Close'] / dfn50['Close'].shift(1))

dfb500['PercentChange'] = dfb500['Close'].pct_change()
dfb500['LogReturn'] = np.log(dfb500['Close'] / dfb500['Close'].shift(1))
```

# The GARCH(1,1) Model (implemented for NIFTY 50)

- The equation for GARCH(1,1) is
$$\sigma_n^2 = \gamma V_L + \alpha u_{n-1}^2 + \beta\sigma_{n-1}^2$$

- Here $V_L$ denotes the *long-run average variance rate* (and GARCH(1,1) model is mean reverting to this long term rate).\
The $\gamma V_L$ term is typically denoted by variable $\omega$.\
The equation above can be thought of as a weighted average with the sum of the weights being 1.
$$\gamma + \alpha + \beta = 1$$

- The $(1,1)$ in GARCH(1,1) indicates that $\sigma_n^2$ is based on the most recent observation of $u^2$ and the most recent estimate of the variance rate.

- If we let $v_i$ denote $\sigma_i^2$, then the *maximum likelihood method* for estimating $\sigma_n$ reduces to maximizing
$$\sum_{i=1}^{m}\left(\ln(v_i) - \frac{u_i^2}{v_i}\right)$$

**Remarks about implementation**
- We have a *minimize* function in scipy which we use (and don't have a maximize function).

- We write a function *garch_optimization_func* that takes as variables $\omega, \alpha, \beta$ and minimizes
$$-\sum_{i=1}^{m}\left(\ln(v_i) - \frac{u_i^2}{v_i}\right)$$

- The routine works well when the variables are structured to be in similar search spaces.\
So, we adjust $\omega$ by a factor of $10^5$ so that it is close to values of $\alpha$ and $\beta$. 

```python
def optimization_func(p):
    l = df50.shape[0]
    omega, alpha, beta = p[0], p[1], p[2]
    v = np.zeros(l)
    u = df50['LogReturn']**2
    s = 0
    df50['DailyVolatility']=np.sqrt(v)
    for i in range(l):
        if i>0:
            if i==1:
                v[i] = omega/((10**5)*(1-alpha-beta)) #Set to estimate of long term average rate when there are no prior values
            else:
                v[i] = omega/(10**5) + alpha*(u[i-1]) + beta*v[i-1]
                s = s - np.log(v[i]) - (u[i]/v[i])
    if alpha+beta<=1:
        s = (-1)*s
        df50['DailyVolatility']=np.sqrt(v)
        return s
    else:
        return 0
```

- We find the optimal values of $\omega, \alpha, \beta$ and calculate $\gamma$, $V_L$.\
We add a 'DailyVolatility' column to our data frame with the values of $\sigma_n$ found by our model.

- We require $\alpha + \beta$ to be less than $1$ for it to be a *mean reverting* model rather than a *mean fleeing* model.\
If the optimal value for $\omega$ is positive, then it is a consistency check to use GARCH(1,1) model.\
Otherwise, we could have used EWMA model instead.

```python
omega, alpha, beta = 0.1, 0.05, 0.9 #Some initial parameters
vals = scipy.optimize.minimize(optimization_func, x0=[omega,alpha,beta], bounds = ((.001,1),(.001,1),(.001,1)))
optimization_func([vals.x[0],vals.x[1],vals.x[2]])
omega, alpha, beta = vals.x[0]/(10**5),vals.x[1],vals.x[2]
gamma = 1 - alpha - beta
v_l = omega/(1-alpha-beta)
optimization_func([vals.x[0],vals.x[1],vals.x[2]]) #Re-run the optimization function with optimal values to update 'DailyVolatility'
print("Long Term Volatility  :",v_l)
print("Optimal value of gamma:",gamma)
print("Optimal value of omega:",omega)
print("OPtimal value of alpha:",alpha)
print("Optimal value of  beta:",beta)
```

# Alternative estimate of volatilities
- One could ignore the historical data from distant past and only consider the recent $m$ days of data to estimate volatility.\
The expected change in a variable is very small when compared with the standard deviation of changes.\
So, we assume that the mean of the daily log return over the previous $n$ days is 0.


- We do a rolling volatility calculation on the previous $m$ days.\
This makes it a maximum likelihood estimate instead of an unbiased estimate but it simplifies calculations.\
We have initialized $m$ as $21$ which is roughly the number of trading days in a month.

```python
df50['RollingVolatility'] = np.sqrt((df50.LogReturn**2).rolling(m).mean())
df50.head()#The first few rows of the data frame for NIFTY 50
```

```python
df50.tail()#The last few rows of the data frame for NIFTY 50
```

- We first plot the volatilities calculated on a rolling basis from the previous $m$ days.

- We then superimpose and plot the volatilities from GARCH(1,1) model to show how it is a finer estimate.

```python
plt.figure(figsize=(12, 8))
plt.plot_date(df50.Date, df50.RollingVolatility, fmt='-', label='NIFTY 50 Rolling Volatility estimate');
plt.legend();
```

```python
plt.figure(figsize=(12, 8))
plt.plot_date(df50.Date, df50.RollingVolatility, fmt='-', label='NIFTY 50 Rolling Volatility estimate');
plt.plot_date(df50.Date, df50.DailyVolatility, fmt='-', label='NIFTY 50 GARCH(1,1) Volatility estimate');
plt.legend();
```

- By comparing with rolling estimates, the estimates from GARCH(1,1) model are satisifactory with back testing.\
The comparison also shows it is particularly good during stress testing when there were periods with extreme changes in volatility.

- The *Ljung-Box Statistic* is a scientific test to check if our implementation of GARCH(1,1) model explains the data.\
In particular, we calculate the autocorrelation for $u_i^2$, $u_i^2/\sigma_i^2$ (and for rolling volatility respectively) with lag $m$.\
A significant reduction in this statistic for $u_i^2/\sigma_i^2$ suggests that autocorrelation has been largely removed by GARCH model.

**Remarks about implementation**
- We use the inbuilt *acorr_ljungbox* function from *statsmodels* to calculate the Ljung-Box Statistic.

- This serves as a robust check for our earlier implementation.

- We only compare the statistic for a particular lag value but it could be done for a larger set of values leading to the same conclusion.

```python
lb_u = sm.stats.diagnostic.acorr_ljungbox((df50['PercentChange']**2)[1:],lags=m)
lb_g = sm.stats.diagnostic.acorr_ljungbox((df50['PercentChange']**2/df50['DailyVolatility']**2)[1:],lags=m)
lb_r = sm.stats.diagnostic.acorr_ljungbox((df50['PercentChange']**2/df50['RollingVolatility']**2)[m:],lags=m)
print("The Ljung-Box statistic for (u_i)^2            :",lb_u.lb_stat.iloc[-1])
print("The Ljung-Box statistic for (u_i)^2/(sigma_i)^2:",lb_g.lb_stat.iloc[-1])
print("The Ljung-Box statistic from rolling volatility:",lb_r.lb_stat.iloc[-1])
```

We see from this test that our implementation of GARCH(1,1) model has the lowest Ljung-Box statistic and hence explains the data.

# Using GARCH(1,1) to forecast future volatility

- Repeatedly using the equation for GARCH(1,1) model, we get that
$$\mathbb{E}(\sigma_{n+t}^2) = V_L + (\alpha + \beta)^t(\sigma_n^2 - V_L)$$

- We choose a value of $t$ and predict the value of volatility $t$ days in the future.

```python
t = 1000
z = np.zeros(t)
for i in range(t):
    z[i] = v_l + ((alpha + beta)**i)*(df50.DailyVolatility.iloc[-1] - v_l)

print(df50.DailyVolatility.iloc[-1], v_l)
plt.plot(list(range(t)),z)
plt.xlabel('Number of days from Aug 30, 2024', fontsize = 15)
plt.ylabel('Estimate of volatility of NIFTY50', fontsize = 15)
plt.show()
```

```python
def optimization_func(p):
    l = dfb500.shape[0]
    omega, alpha, beta = p[0], p[1], p[2]
    v = np.zeros(l)
    u = dfb500['LogReturn']**2
    s = 0
    dfb500['DailyVolatility']=np.sqrt(v)
    for i in range(l):
        if i>0:
            if i==1:
                v[i] = omega/((10**5)*(1-alpha-beta)) #Set to estimate of long term average rate when there are no prior values
            else:
                v[i] = omega/(10**5) + alpha*(u[i-1]) + beta*v[i-1]
                s = s - np.log(v[i]) - (u[i]/v[i])
    if alpha+beta<=1:
        s = (-1)*s
        dfb500['DailyVolatility']=np.sqrt(v)
        return s
    else:
        return 0
```

```python
omega, alpha, beta = 0.1, 0.05, 0.9 #Some initial parameters
vals = scipy.optimize.minimize(optimization_func, x0=[omega,alpha,beta], bounds = ((.001,1),(.001,1),(.001,1)))
optimization_func([vals.x[0],vals.x[1],vals.x[2]])
omega, alpha, beta = vals.x[0]/(10**5),vals.x[1],vals.x[2]
gamma = 1 - alpha - beta
v_l = omega/(1-alpha-beta)
optimization_func([vals.x[0],vals.x[1],vals.x[2]]) #Re-run the optimization function with optimal values to update 'DailyVolatility'
print("Long Term Volatility  :",v_l)
print("Optimal value of gamma:",gamma)
print("Optimal value of omega:",omega)
print("OPtimal value of alpha:",alpha)
print("Optimal value of  beta:",beta)
```
